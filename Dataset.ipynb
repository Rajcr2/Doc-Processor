```python
import os
import pandas as pd
from huggingface_hub import HfApi, hf_hub_download

df = pd.read_csv("hf://datasets/theatticusproject/cuad/CUAD_v1/master_clauses.csv")
subset = set(df["Filename"].head(50))

cuad_data = HfApi().list_repo_files("theatticusproject/cuad", repo_type="dataset")
pdf_paths = [f for f in cuad_data if f.endswith(".pdf") and os.path.basename(f) in subset]

os.makedirs("contracts", exist_ok=True)

for path in pdf_paths:
    local = hf_hub_download(repo_id="theatticusproject/cuad", filename=path, repo_type="dataset")
    os.rename(local, f"contracts/{os.path.basename(path)}")
    print(f"✓ {os.path.basename(path)}")

print(f"\n{len(os.listdir('contracts'))} Contracts.")
```
